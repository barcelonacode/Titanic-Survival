In the above we dropped the Nan values but we lost a significant amount of data so we will replace the Nan values with the mean or median of the particular column

### checking for which columns have Nan Values in our considered Features

```python
df['Sex'].isnull().sum() # No Nan Values
```

```python
df['Age'].isnull().sum() # 177 Nan Values hence it is not good to drop these Nan Values
```

```python
df['Fare'].isnull().sum()
```

```python
df['Embarked'].isnull().sum()
```

### Filling the Null Values with the median of the column

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'].isnull().sum() # So now no Nan Values
```
